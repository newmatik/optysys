```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.patches as patches
import pandas as pd

import sys
```

```python
log_file = []
with open('train_log.txt') as f:
    lines = f.readlines()

for line in lines:
    log_file.append(line.split())
```

```python
train_loss = []
for i in log_file:
    train_loss.append(float(i[3]))

train_loss[1]=10#train_loss[1]/100
idx = 0
plt.plot(range(idx+1,len(train_loss)+1),train_loss[idx:])
plt.scatter(np.argmin(train_loss)+1,np.min(train_loss),c='r')
```
